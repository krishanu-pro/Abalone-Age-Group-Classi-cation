### Adaboost for mean imputation

```python
import pandas
data = pandas.read_csv("mean_data.csv")
cl=data.columns
X=data[cl[1:len(cl)-1]]
y=data['Rings']
print(X)
print(y)
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn import datasets
# Import train_test_split function
from sklearn.model_selection import train_test_split
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
from sklearn.model_selection import train_test_split
# dividing X, y into train and test data 
X_train, X_test, y_train, y_test = train_test_split(X, y,  train_size=0.7, test_size=0.3, random_state=123) 
```

```python
# Create adaboost classifer object
abc = AdaBoostClassifier()
from pprint import pprint
pprint(abc.get_params())
```

```python
# Train Adaboost Classifer
model = abc.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = model.predict(X_test)
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV 
ada=AdaBoostClassifier()
search_grid={'n_estimators':[50,100,500,1000,2000],'learning_rate':[.001,0.01,.1]}
search=GridSearchCV(estimator=ada,param_grid=search_grid,scoring='accuracy',n_jobs=1,cv=3)
```

```python
search.fit(X_train,y_train)
search.best_params_
```

```python
train_pred = search.predict(X_train)
accuracy = search.score(X_test, y_test) 
print(accuracy) 
    
```
